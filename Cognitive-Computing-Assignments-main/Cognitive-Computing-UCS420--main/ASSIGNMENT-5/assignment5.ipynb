```python
import numpy as np
```

```python
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print(gfg)

sum = np.sum(gfg)
print("Sum of all elements:",sum)

row_wise_sum = np.sum(gfg, axis=1)
print("Sum of elements row-wise:\n", row_wise_sum)

column_wise_sum = np.sum(gfg, axis=0)
print("Sum of elements column-wise:", column_wise_sum)
```

```python
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
print("Sorted array:", sorted_array)

indices = np.argsort(array)
print("Indices of sorted array:", indices)

smallest_4 = np.sort(array)[:4]
print("4 smallest elements:", smallest_4)

largest_5 = np.sort(array)[-5:]
print("5 largest elements:", largest_5)

array2= np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array2[array2 == array2.astype(int)]
print("Integer elements:", integer_elements)

float_elements = array2[array2 == array2.astype(float)]
print("Float elements:", float_elements)

float_elements = array2[array2 != array2.astype(int)]
print("Float elements:", float_elements)
```

```python
X = ord('E') + ord('S')
sales = np.array([X, X + 50, X + 100, X + 150, X + 200])
print("Sales dataset:", sales)

tax_rate = ((X % 5) + 5) / 100
sales_after_tax = sales * (1 + tax_rate)
print("Sales after tax:", sales_after_tax)

discounted_sales=[]
for sale in sales:
    if sale < X + 100:
        discounted_sales.append(sale * 0.95)
    elif sale >= X + 100:
        discounted_sales.append(sale * 0.90) 
print("Sales after discounting:",discounted_sales)

weeks = 3
weekly_sales = np.vstack([sales] * weeks)
growth_factor = 1.02 ** np.arange(weeks)[:, None]
weekly_sales_adjusted = weekly_sales * growth_factor

print("Adjusted sales for 3 weeks:\n", weekly_sales_adjusted)
```

```python
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y1 = x**2               
y2 = np.sin(x)           
y3 = np.exp(x)          
y4 = np.log(np.abs(x) + 1)
```

```python
plt.plot(x, y1, label="y = x^2", color='blue')
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```

```python
plt.plot(x, y2, label="y = sin(x)", color='red')
plt.title("y = sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```

```python
plt.plot(x, y3, label="y = e^x", color='lime')
plt.title("y = e^x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```

```python
plt.plot(x, y4, label="y = log(|x| + 1)", color='cyan')
plt.title("y = log(|x| + 1)")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)

plt.show()
```
